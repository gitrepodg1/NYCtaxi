## Open question A:  when can you use mean as measure of central tendency

Let's start by finding the number of hack_licenses (drivers) and how much they earn each day

```python
def show_distribution(df,xlabel,bins=500):
    df.hist(bins=bins);       #for each driver, their average daily income averaged across the 30 days
    plt.xlabel(xlabel);
    plt.ylabel('Count');
    plt.title("Distribution of "+xlabel);
    print('mean  =',df.mean())
    print('median=',df.median())
    #print('mode  =',df.mode())
    print('stddev=',df.std())
    print('skew  =',df.skew())
```

```python
show_distribution(df_remove_outlier_latitudelongitude[' fare_amount'],'fare_amount')
```

```python
show_distribution(df_remove_outlier_latitudelongitude[df_remove_outlier_latitudelongitude[' fare_amount']<100][' fare_amount'],'fare_amount')
```

```python
show_distribution(df_remove_outlier_latitudelongitude[' trip_time_in_secs'],'trip_time_in_secs',150)
```

```python
show_distribution(df_remove_outlier_latitudelongitude[df_remove_outlier_latitudelongitude[' trip_time_in_secs']<2000][' trip_time_in_secs'],'trip_time_in_secs',1450)
```
